```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

df_stars = pd.read_csv("Stars.csv")
df_adult = pd.read_csv("adult.csv")
```

1. Ulangi langkah pada kelas praktikum untuk dataset Star dan Adult. Masing masing gunakan perbandingan 0.2, 0.3, 0.4. Tuliskan akurasinya

Star

```python
df_stars.drop(['Unnamed: 0'],axis='columns',inplace=True) #menghapus kolom 'Unnamed: 0'
df_stars.head()
```

```python
encoder = LabelEncoder()
df_stars['Star category'] = encoder.fit_transform(df_stars['Star category']) #mengonversi Star Category menjadi biner
df_stars.head()
```

```python
features = df_stars.drop('Star category',axis='columns') #hapus star category di data features
target = df_stars['Star category'] #jadikan star category data target
features.info()
```

```python
model = GaussianNB() #model naive bayes

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(features, target, test_size=0.2) #20%
model.fit(X_train1, Y_train1)
acc_1 = accuracy_score(Y_test1, model.predict(X_test1))

X_train2, X_test2, Y_train2, Y_test2 = train_test_split(features, target, test_size=0.3) #30%
model.fit(X_train2, Y_train2)
acc_2 = accuracy_score(Y_test2, model.predict(X_test2))

X_train3, X_test3, Y_train3, Y_test3 = train_test_split(features, target, test_size=0.4) #40%
model.fit(X_train3, Y_train3)
acc_3 = accuracy_score(Y_test3, model.predict(X_test3))

print(f"accuracy_score dari data test sebesar 20%:", acc_1)
print(f"accuracy_score dari data test sebesar 30%:", acc_2)
print(f"accuracy_score dari data test sebesar 40%:", acc_3)
```

2. Untuk masing masing dataset carilah subset kombinasi dari atribut yang memberi hasil klasifikasi lebih baik dari data asli

```python
df_stars = pd.read_csv("Stars.csv")
df_stars.drop(['Unnamed: 0'], axis='columns', inplace=True) #menghapus kolom 'Unnamed: 0'

labelencoder = LabelEncoder()
df_stars['Star category'] = labelencoder.fit_transform(df_stars['Star category']) #mengonversi kolom Star Category

features = df_stars.drop('Star category', axis='columns')
target = df_stars['Star category'] #membagi data feature dan target

model = GaussianNB()
sfs = SFS(model,
          k_features=(1, len(features.columns)),
          forward=True,
          floating=False,
          scoring='accuracy',
          cv=5) #membuat Sequential Forward Selection

sfs.fit(features, target) #membuat model SFS

selected_features = list(sfs.k_feature_names_) #mendapatkan nama feature

model.fit(features[selected_features], target) #menyeleksi subset dari features untuk akurasi model yang lebih baik

accuracy = accuracy_score(target, model.predict(features[selected_features])) #evaluasi model

print("Features terpilih:", selected_features)
print("accuracy_score didapat dari features terpilih:", accuracy)
```

```python
features.drop(['Temperature (K)', 'Luminosity (L/Lo)'],axis='columns',inplace=True)  #menghapus kolom yang berkorelasi rendah
features.head()
```

```python
model = GaussianNB() #model naive bayes

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(features, target, test_size=0.2) #20%
model.fit(X_train1, Y_train1)
acc_1 = accuracy_score(Y_test1, model.predict(X_test1))

X_train2, X_test2, Y_train2, Y_test2 = train_test_split(features, target, test_size=0.3) #30%
model.fit(X_train2, Y_train2)
acc_2 = accuracy_score(Y_test2, model.predict(X_test2))

X_train3, X_test3, Y_train3, Y_test3 = train_test_split(features, target, test_size=0.4) #40%
model.fit(X_train3, Y_train3)
acc_3 = accuracy_score(Y_test3, model.predict(X_test3))

print(f"accuracy_score dari data test sebesar 20%:", acc_1)
print(f"accuracy_score dari data test sebesar 30%:", acc_2)
print(f"accuracy_score dari data test sebesar 40%:", acc_3)
```

```python
df_adult = pd.read_csv("adult.csv")
dummies = pd.get_dummies(df_adult['race']) #mengonversi nilai pada kolom ras dari kategorial menjadi binary
df_adult = pd.concat([df_adult,dummies],axis='columns') #menggabungkan hasil konversi dengan dataframe utama

df_adult.drop(['fnlwgt', 'education', 'race', 'capital-gain', 'capital-loss',
            'hours-per-week', 'native-country', 'Amer-Indian-Eskimo',
            'Asian-Pac-Islander', 'Other', 'White'],axis='columns',inplace=True) #menghapus kolom tidak perlu

adult_columns = ['workclass', 'marital-status', 'occupation', 'relationship', 'gender', 'income', 'Black']
for col in adult_columns:
  df_adult[col] = labelencoder.fit_transform(df_adult[col])  #memberi label pada setiap nilai kategorikal

features = df_adult.drop('Black',axis='columns') #menghapus black race dari features
target = df_adult['Black'] #menjadikan black race target

model = GaussianNB() #model naive bayes
sfs = SFS(model,
          k_features=(1, len(features.columns)),
          forward=True,
          floating=False,
          scoring='accuracy',
          cv=5) #membuat SFS model

sfs.fit(features, target) #membuat model SFS

selected_features = list(sfs.k_feature_names_) #mendapatkan nama feature

model.fit(features[selected_features], target) #menyeleksi subset dari features untuk akurasi model yang lebih baik

accuracy = accuracy_score(target, model.predict(features[selected_features])) #evaluasi model

print("Features terpilih:", selected_features)
print("accuracy_score didapat dari features terpilih:", accuracy)
```

```python
model = GaussianNB() #model naive bayes

X_train1, X_test1, Y_train1, Y_test1 = train_test_split(features, target, test_size=0.2) #20%
model.fit(X_train1, Y_train1)
acc_1 = accuracy_score(Y_test1, model.predict(X_test1))

X_train2, X_test2, Y_train2, Y_test2 = train_test_split(features, target, test_size=0.3) #30%
model.fit(X_train2, Y_train2)
acc_2 = accuracy_score(Y_test2, model.predict(X_test2))

X_train3, X_test3, Y_train3, Y_test3 = train_test_split(features, target, test_size=0.4) #40%
model.fit(X_train3, Y_train3)
acc_3 = accuracy_score(Y_test3, model.predict(X_test3))

print(f"accuracy_score dari data test sebesar 20%:", acc_1)
print(f"accuracy_score dari data test sebesar 30%:", acc_2)
print(f"accuracy_score dari data test sebesar 40%:", acc_3)
```
