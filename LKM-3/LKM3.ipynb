```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io
```

1. Masukkan iris.csv ke dataframe. Normalisasikan setiap atribut iris ke dataframe menggunakan normalisasi Min-Max sehingga setiap nilai atribut memiliki range antara 0 sampai 1. Dengan Vbaru adalah nilai data disuatu sel setelah normalisasi, Vlama adalah nilai data asli sebelum normalisasi, Vmax adalah nilai max dari kolom data tersebut, dan Vmin adalah nilai min dari kolom data tersebut.

```python
df_iris = pd.read_csv('iris.csv')
df_iris.head()
```

```python
sepalWidth_max,sepalWidth_min = df_iris['sepal.width'].max(), df_iris['sepal.width'].min()
petalWidth_max, petalWidth_min = df_iris['petal.width'].max(), df_iris['petal.width'].min()
sepalLength_max, sepalLength_min = df_iris['sepal.length'].max(), df_iris['sepal.length'].min()
petalLength_max, petalLength_min = df_iris['petal.length'].max(), df_iris['petal.length'].min()

for i in df_iris.index :
  df_iris.loc[i, 'sepal.width'] = (df_iris.loc[i, 'sepal.width']-sepalWidth_min)/(sepalWidth_max-sepalWidth_min)
  df_iris.loc[i, 'petal.width'] = (df_iris.loc[i, 'petal.width']-petalWidth_min)/(petalWidth_max-petalWidth_min)
  df_iris.loc[i, 'sepal.length'] = (df_iris.loc[i, 'sepal.length']-sepalLength_min)/(sepalLength_max-sepalLength_min)
  df_iris.loc[i, 'petal.length'] = (df_iris.loc[i, 'petal.length']-petalLength_min)/(petalLength_max-petalLength_min)

print(df_iris)
```

2. Carilah centroid dari data yang telah dinormalisasi. Centroid dihitung dari rata-rata seluruh data suatu kolom.

```python
print ('SETOSA')
splCentroid_setosa = df_iris[df_iris.variety == 'Setosa']['sepal.length'].mean()
print (splCentroid_setosa)
spwCentroid_setosa = df_iris[df_iris.variety == 'Setosa']['sepal.width'].mean()
print (spwCentroid_setosa)
ptlCentroid_setosa = df_iris[df_iris.variety == 'Setosa']['petal.length'].mean()
print (ptlCentroid_setosa)
ptwCentroid_setosa = df_iris[df_iris.variety == 'Setosa']['petal.width'].mean()
print (ptwCentroid_setosa)

print ('\nVERSICOLOR')
splCentroid_versicolor = df_iris[df_iris.variety == 'Versicolor']['sepal.length'].mean()
print (splCentroid_versicolor)
spwCentroid_versicolor = df_iris[df_iris.variety == 'Versicolor']['sepal.width'].mean()
print (spwCentroid_versicolor)
ptlCentroid_versicolor = df_iris[df_iris.variety == 'Versicolor']['petal.length'].mean()
print (ptlCentroid_versicolor)
ptwCentroid_versicolor = df_iris[df_iris.variety == 'Versicolor']['petal.width'].mean()
print (ptwCentroid_versicolor)

print ('\nVIRGINICA')
splCentroid_virginica = df_iris[df_iris.variety == 'Virginica']['sepal.length'].mean()
print (splCentroid_virginica)
spwCentroid_virginica = df_iris[df_iris.variety == 'Virginica']['sepal.width'].mean()
print (spwCentroid_virginica)
ptlCentroid_virginica = df_iris[df_iris.variety == 'Virginica']['petal.length'].mean()
print (ptlCentroid_virginica)
ptwCentroid_virginica = df_iris[df_iris.variety == 'Virginica']['petal.width'].mean()
print (ptwCentroid_virginica)
```

3. Masukkan iristesting.csv ke dataframe. Normalisasikan iristesting.csv menggunakan normalisasi Min-Max.

```python
df_irisTest = pd.read_csv ('iristesting.csv')
df_irisTest.head()
```

```python
sepalWidth_max,sepalWidth_min = df_irisTest['sepal.width'].max(), df_irisTest['sepal.width'].min()
petalWidth_max, petalWidth_min = df_irisTest['petal.width'].max(), df_irisTest['petal.width'].min()
sepalLength_max, sepalLength_min = df_irisTest['sepal.length'].max(), df_irisTest['sepal.length'].min()
petalLength_max, petalLength_min = df_irisTest['petal.length'].max(), df_irisTest['petal.length'].min()

for i in df_irisTest.index :
  df_irisTest.loc[i, 'sepal.width'] = (df_irisTest.loc[i, 'sepal.width']-sepalWidth_min)/(sepalWidth_max-sepalWidth_min)
  df_irisTest.loc[i, 'petal.width'] = (df_irisTest.loc[i, 'petal.width']-petalWidth_min)/(petalWidth_max-petalWidth_min)
  df_irisTest.loc[i, 'sepal.length'] = (df_irisTest.loc[i, 'sepal.length']-sepalLength_min)/(sepalLength_max-sepalLength_min)
  df_irisTest.loc[i, 'petal.length'] = (df_irisTest.loc[i, 'petal.length']-petalLength_min)/(petalLength_max-petalLength_min)

print(df_irisTest)
```

4. Prediksilah kelas data di iristesting.csv menggunakan kriteria kedekatan distance antara setiap data di iristesting.csv ke centroid masing2 class di iris. Gunakan euclidean distance.

```python
setosa, versicolor, virginica = [], [], []

for i in df_irisTest.index :
  if (df_irisTest.variety[i] == 'Setosa'):
    gap = ((df_irisTest['sepal.length'][i] - splCentroid_setosa)**2 + (df_irisTest['sepal.width'][i] - spwCentroid_setosa)**2 + (df_irisTest['petal.length'][i] - ptlCentroid_setosa)**2 + (df_irisTest['petal.width'][i] - ptwCentroid_setosa)**2 )**1/2
    setosa.append(gap)
  elif (df_irisTest.variety[i] == 'Versicolor'):
    gap = ((df_irisTest['sepal.length'][i] - splCentroid_versicolor)**2 + (df_irisTest['sepal.width'][i] - spwCentroid_versicolor)**2 + (df_irisTest['petal.length'][i] - ptlCentroid_versicolor)**2 + (df_irisTest['petal.width'][i] - ptwCentroid_versicolor)**2)**1/2
    versicolor.append(gap)
  elif (df_irisTest.variety[i] == 'Virginica'):
    gap = ((df_irisTest['sepal.length'][i] - splCentroid_virginica)**2 + (df_irisTest['sepal.width'][i] - spwCentroid_virginica)**2 + (df_irisTest['petal.length'][i] - ptlCentroid_virginica)**2 + (df_irisTest['petal.width'][i] - ptwCentroid_virginica)**2)**1/2
    virginica.append(gap)

result = [setosa, versicolor, virginica]

for i in result :
  print(i)
```

5. Hitunglah berapa persen klasifikasi yang benar (akurasi) saat memprediksi class 

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
a = df_irisTest.iloc[:, :-1].values
print(a)
b = df_irisTest.iloc[:, 4].values
a_train, a_test, b_train, b_test = train_test_split(a, b, test_size=0.20)
classifier = KNeighborsClassifier (n_neighbors=5)
classifier.fit(a_train, b_train)
pred = classifier.predict(a_test)

print ('Predict')
Predict = classifier.predict([[0.60, 0.466667, 0.826087, 0.695652]])
print ('\nPredict Result', Predict)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(accuracy_score(b_test, pred))
```
