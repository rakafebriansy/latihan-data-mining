4. Masukkan dataset iris sebagai array 2D di numpy. Pisahkan setiap kelas di array yang berbeda : setosa, versicolor, virginica, dan unknown

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Memasukkan dataset iris sebagai array 2D
iris = pd.read_csv("iris_dataset.csv")
array_iris = iris.to_numpy()

# Memisahkan kelas di array yang berbeda
setosa = array_iris[array_iris[:, -1] == 'Setosa']
versicolor = array_iris[array_iris[:, -1] == 'Versicolor']
virginica = array_iris[array_iris[:, -1] == 'Verginica']
unknown = array_iris[array_iris[:, -1] == 'unknown']
print(setosa)
```

```python
x = setosa[:,4]
y = setosa[:,1]
plt.scatter(x,y, color = 'red')

x = versicolor[:,4]
y = versicolor[:,1]
plt.scatter (x, y, color ='green')

x = virginica[:,4]
y = virginica[:,1]
plt.scatter (x, y, color ='blue')

x = unknown[:,4]
y = unknown[:,1]
plt.scatter (x, y, color = 'yellow')
plt.title("Sepal length dan petal width")
plt.show()
```

8. Buatlah paralel coordinat plot dari data yang berkelas setosa, versicolor, dan virginica. Berikan warna merah untuk setosa, hijau untuk versicolor, biru untuk virginica. Hint gunakan perintah plot dan operasi slicing pada numpy untuk mempermudah tugas anda

```python
# Memisahkan data untuk coordinat plot
setosa_data = array_iris[array_iris[:, -1] == 'Setosa', 1:5]
versicolor_data = array_iris[array_iris[:, -1] == 'Versicolor', 1:5]
virginica_data = array_iris[array_iris[:, -1] == 'Verginica', 1:5]

# Membuat plot
plt.plot(setosa_data.T, color ='red')
plt.plot(versicolor_data.T, color ='green')
plt.plot(virginica_data.T, color ='blue')

# Keterangan Data
plt.title("Iris")
plt.xticks(np.arange(4), ['Petal_width', 'Petal_length', 'Sepal_width', 'Sepal_length'])
plt.xlabel("Atribut")

plt.show()
```
